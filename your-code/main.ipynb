```python
# Method 1 - matplotlib
# your code here
gender = df.groupby("Gender").count()[["Name"]]
gender = gender.reset_index().rename(columns={"Name":"cnt"})
gender
```

```python
fem = gender["cnt"][0]/ gender["cnt"].sum()
mal = gender["cnt"][1]/ gender["cnt"].sum()
sizes = [fem, mal]
plt.pie(sizes, explode = (0,0.1), labels=["female", "male"],autopct='%1.1f%%', shadow = 2, startangle=90)
plt.show()
```
